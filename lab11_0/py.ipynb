```python
Вектори, створення та робота з ними

import numpy as np

print("Все можна зробити вручну за допомогою методу np.array, передавши йому набір значень у вигляді List:")

# Створимо рядковий вектор (один рядок, кілька стовпців)
vector_row = np.array([[1, -5, 3, 2, 4]])

# Створимо стовпцевий вектор (кілька рядків, один стовпець)
vector_column = np.array([[1], [2], [3], [4]])

# Виведемо рядковий вектор та його розмір
print("Рядковий вектор:\n", vector_row)
print("Розмір рядкового вектора:", vector_row.shape)
# Пояснення:
# Розмір цього масиву (1, 5) означає, що це масив з 1 рядком і 5 стовпцями.

# Виведемо стовпцевий вектор та його розмір
print("\nСтовпцевий вектор:\n", vector_column)
print("Розмір стовпцевого вектора:", vector_column.shape)
# Пояснення:
# Розмір цього масиву (4, 1) означає, що це масив з 4 рядками і 1 стовпцем.

# Додатково можемо вивести кілька корисних фактів
print("\nМаємо можливість перетворювати вектори в одномірні за допомогою reshape:")
reshaped_vector_row = vector_row.reshape(-1)  # Перетворюємо рядковий вектор на одномірний
reshaped_vector_column = vector_column.reshape(-1)  # Перетворюємо стовпцевий вектор на одномірний
print("Рядковий вектор після reshape:", reshaped_vector_row)
print("Стовпцевий вектор після reshape:", reshaped_vector_column)
```

```python
Метод linspace - генерує задану кількістю чисел у певному діапазоні, 50 чисел за замовчуванням

import numpy as np

# Параметри генерації
start = 5
size_of_vec = 100

# Генерація масиву з використанням linspace
a = np.linspace(start, size_of_vec, num=50)  # Генеруємо 50 чисел в діапазоні від start до size_of_vec
print(f"Згенерували {len(a)} чисел в діапазоні від {a[0]} до {a[-1]}")
# Пояснення: np.linspace генерує лінійно розподілені числа між start та size_of_vec включно.
# Ми виводимо перше і останнє значення масиву, а також кількість чисел.
homework("Виведіть кожне друге число згенерованого діапазону. Тут буде потрібен цикл :)")

# Зменшимо кількість чисел до 5
a = np.linspace(start, size_of_vec, 5)
print("Згенеровано 5 чисел:", a)
# Пояснення: при використанні 5 чисел np.linspace автоматично визначить крок, щоб рівномірно розподілити ці числа.
homework("Збільшіть кількість чисел до 100. Виведіть перших та останніх 10.")

# Як бачимо, перше та останнє значення завжди входять у згенерований діапазон.
# Це означає, що крок вираховується автоматично таким чином, щоб включити ці значення.
print("""
Як бачимо, перше та останнє значення завжди входять у згенерований діапазон. 
Відповідно, при цьому вираховується крок.
Це можна змінити, виключивши останнє значення з визначення діапазону.
""")

# Якщо ми не хочемо включати останнє значення, використовуємо параметр `endpoint=False`
print("Масив без включення останнього значення:")
print(np.linspace(start, size_of_vec, 5, endpoint=False))
homework("Спробуйте самі порахувати крок.")

# Отримуємо крок за допомогою параметра retstep
_, sp_1 = np.linspace(start, size_of_vec, 5, endpoint=True, retstep=True)
_, sp_2 = np.linspace(start, size_of_vec, 5, endpoint=False, retstep=True)

print(f"Крок для endpoint=True: {sp_1}")
print(f"Крок для endpoint=False: {sp_2}")
# Пояснення: ми отримуємо крок, використовуючи параметр retstep, який дає значення кроку для кожного випадку.
homework("Чому маємо такий крок?")
```
